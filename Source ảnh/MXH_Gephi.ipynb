```python
import networkx as nx

G = nx.read_edgelist("facebook_combined_cleaned.txt", nodetype=int)
print("Nodes:", G.number_of_nodes())
print("Edges:", G.number_of_edges())
```

# **T√≠nh community**

```python
partition = community_louvain.best_partition(G)

# partition d·∫°ng {node: community_id}
num_communities = len(set(partition.values()))
print("S·ªë community:", num_communities)
```

# **T√≠nh centrality**

```python
# Degree centrality
degree_dict = nx.degree_centrality(G)

# Hi·ªÉn th·ªã 10 node c√≥ degree cao nh·∫•t
print("üîπ Top 10 Degree Centrality:")
for node, val in sorted(degree_dict.items(), key=lambda x: x[1], reverse=True)[:10]:
    print(f"Node {node} ‚Üí {val:.5f}")


# Betweenness centrality
betweenness_dict = nx.betweenness_centrality(G, normalized=True)

# Hi·ªÉn th·ªã 10 node c√≥ Betweenness cao nh·∫•t
print("\nüîπ Top 10 Betweenness Centrality:")
for node, val in sorted(betweenness_dict.items(), key=lambda x: x[1], reverse=True)[:10]:
    print(f"Node {node} ‚Üí {val:.5f}")

```

# **G√°n thu·ªôc t√≠nh v√†o node**

```python
nx.set_node_attributes(G, partition, "community")
nx.set_node_attributes(G, degree_dict, "degree_centrality")
nx.set_node_attributes(G, betweenness_dict, "betweenness_centrality")
```
